```python
import csv

# Specify the path to the input CSV file
input_file = '/content/data/lex_avgs/lexicon_output.csv'

# Specify the path to the output CSV file
output_file = '/content/data/lex_avgs/lexicon_output_fixed.csv'

# Open the input and output files
with open(input_file, 'r') as f_input, open(output_file, 'w', newline='') as f_output:
    # Create a CSV reader and writer objects
    reader = csv.reader(f_input)
    writer = csv.writer(f_output)

    # Write the first row (header) to the output file
    writer.writerow(next(reader))

    # Iterate over the remaining rows
    for row in reader:
        # Check if the first column starts with 'men' or 'women'
        if row[0].startswith(('men', 'women')):
            # Write the row to the output file
            writer.writerow(row)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data from the CSV file
f_path = "/content/data/log_odds/log_odds.txt"
data = pd.read_csv(f_path, names=['word', 'correlation degree'])

# Sort the data by 'correlation degree' in descending order
data_sorted = data.sort_values(by='correlation degree', ascending=False)

# Select the top 30 words
top_30_words = data_sorted.head(30)

# Set the style for the plot
sns.set_style("whitegrid")

# Create the figure and axis objects
plt.figure(figsize=(80, 35))

# Plot the data using sns.barplot
ax = sns.barplot(x='word', y='correlation degree', data=top_30_words, palette=sns.color_palette("rocket"))

# Customize the plot
ax.tick_params(labelsize=35)
ax.set_xlabel('Word', fontsize=50)
ax.set_ylabel('Correlation Degree', fontsize=50)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize=50)
ax.set_title("GPT-3/InstructGPT Changed Genders Male Highest 30 Correlating Adjectives & Verbs", fontsize=60)

# Save the plot to a file
plt.savefig('GPT-3-Changed-genders-top-words-men.png')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

f_path = "/content/data/lex_avgs/lexicon_output_fixed.csv"
csv = pd.read_csv(f_path, header=0)

# Plot the data
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

sns.barplot(
    x='dimension',
    y='mean',
    hue='category',
    data=csv,
    palette=palette,
    # yerr=[csv['ci_lower'], csv['ci_upper']]  # Specify the lower and upper bounds of the confidence interval
)

plt.xlabel('Dimension')
plt.ylabel('Value')
plt.title('GPT-3 changed genders lexicon analysis')

plt.legend(title='Category')
plt.savefig('/content/data/GPT-3_changed_genders.png')
plt.show()
```
